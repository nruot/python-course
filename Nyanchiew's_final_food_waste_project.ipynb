```python
sorted_df = df.sort_values(by=['Retail estimate (tonnes/year)'], ascending=False)
```

```python
region_waste = df.groupby('Region')['Total Waste'].sum()

# Find the region with the maximum total waste
highest_waste_region = region_waste.idxmax()
highest_waste_amount = region_waste.max()

print(f"The region with the highest food waste is: {highest_waste_region}")
print(f"The amount of food waste in {highest_waste_region} is: {highest_waste_amount}")
```

```python
#graphing food waste and regions
```

```python
region_waste = df.groupby('Region')['Total Waste'].sum()

# Create the bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.bar(region_waste.index, region_waste.values)
plt.xlabel("Region")
plt.ylabel("Total Food Waste")
plt.title("Food Waste by Region")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
#top 10 countries food waste
```

```python
top_10_countries = df.sort_values(by=['Total Waste'], ascending=False).head(10)
```

```python
# Create the bar chart
plt.figure(figsize=(12, 6))  # Adjust figure size if needed
plt.bar(top_10_countries['Country'], top_10_countries['Total Waste'])
plt.xlabel("Country")
plt.ylabel("Total Food Waste")
plt.title("Top 10 Countries with Highest Food Waste")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
#top 10 countries in retail food waste
```

```python
top_10_retail_waste = df.sort_values(by=['Retail estimate (tonnes/year)'], ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))  # Adjust figure size if needed
plt.bar(top_10_retail_waste['Country'], top_10_retail_waste['Retail estimate (tonnes/year)'])
plt.xlabel("Country")
plt.ylabel("Retail Food Waste (tonnes/year)")
plt.title("Top 10 Countries with Highest Retail Food Waste")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
